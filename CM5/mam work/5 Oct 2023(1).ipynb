# Pandas Merging, Joining

```python
#Creating a Employee data Dictionary
data1={
    'Name':['Riya','Pooja', 'nitin','padam'],
    'Age':[25,28,32,22],
    'Address':['Thane','Jaipur','Nagpur','Noida'],
    'Salary':[25000,40000,35000,30000],
    'Qualificatio':['Btech','Bsc','Bcom','Phd']
}
```

```python
#Creating a Employee data Dictionary
data2={
    'Name':['Arti','Sapna', 'nitin','padam'],
    'Age':[27,28,30,22],
    'Address':['Thane','Jaipur','Bhopal','Patna'],
    'Salary':[25000,45000,38000,30000],
    'Qualificatio':['Msc','Bsc','Bcom','MCA']
}
```

```python
#Converting Data into Dataframe
df1=pd.DataFrame(data1,index=[0,1,2,3])
```

```python
#Converting Data into Dataframe
df2=pd.DataFrame(data2, index=[4,5,6,7])
```

```python
print(df1, "\n\n", df2) #
```

```python
#Concatinating the dataframes
a=[df1,df2]
res=pd.concat(a)
```

```python
res #Printing the result
```
